```python
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/SharedTask/SubtaskA_train_Monolingual.csv")
```

```python
df.head()
```

```python
ai_corpus = set()
human_corpus = set()

avgAi = []
avgHumans =  []
for i in range(len(df)):
  text = df['text'].iloc[i].split(" ")

  if df['label'].iloc[i] == 1:
    smallSetAI = set()
    for t in text:
      smallSetAI.add(t)
      ai_corpus.add(t)
      #ai_corpus[t] = ai_corpus.get(t, 0) +1
    avgAi.append(len(smallSetAI))

  else:
    smallSetHumans = set()
    for t in text:
      smallSetHumans.add(t)
      human_corpus.add(t)
      #human_corpus[t] = human_corpus.get(t, 0) +1
    avgHumans.append(len(smallSetHumans))



```
